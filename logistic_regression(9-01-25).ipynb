```python
import matplotlib.pyplot as plt

# Ensure the DataFrame `df` exists with 'Age' and 'Purchased' columns
# Example DataFrame
import pandas as pd
data = {'Age': [22, 25, 47, 52, 46], 'Purchased': [0, 1, 1, 0, 1]}
df = pd.DataFrame(data)

# Scatter plot of Age vs Purchased
plt.figure(figsize=(5, 4))
# Use `c` to map the colors and `cmap` for a colormap
plt.scatter(df['Age'], df['Purchased'], c=df['Purchased'], cmap='bwr', s=100, edgecolor='k')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Purchased (0 = No, 1 = Yes)')
plt.title('Keerthi Project')

# Adding a legend manually
colors = {'No Purchase': 'blue', 'Purchase': 'red'}
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10) for color in colors.values()]
plt.legend(handles, colors.keys(), loc='upper right')

# Adding grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x = df[['Age']]
y = df[['Purchased']]
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=42)
model=LogisticRegression()
model.fit(x_train,y_train)
#prediction 
new_age=[[59]]
prediction = model.predict(new_age)
probability = model.predict_proba(new_age)
print(f"Prediction for Age {new_age[0][0]}: {'will purchase' if prediction[0]==1 else 'will not purchase'}")
print(f"probability: {probability}")
```
